```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#equation of parabola
def function(x):
    return (x+3)**2
```

```python
#derivative of equations
def derivative(x):
    return 2*(x+3)
```

```python
def gradientDecent(starting_point,learning_rate,num_itrations):
    x=starting_point
    history=[]
    for _ in range(num_itrations):
        x-=learning_rate*derivative(x)
        history.append(x)
    return x,history
```

```python
starting_point=2
learning_rate=0.1
num_itrations=100
```

```python
minima,history=gradientDecent(starting_point,learning_rate,num_itrations)
```

```python
print(f"Local minima found at x={minima:.4f},y={function(minima):.4f}")
```

```python
x_values=np.linspace(-6,2,100)
y_values=function(x_values)
plt.plot(x_values,y_values,label='y=(x+3)^2')
plt.scatter(history,function(np.array(history)),color='red',label='Decent path')
plt.title("Gradient decent at Local minima")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```
